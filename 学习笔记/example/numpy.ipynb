```python
# 图片处理
from PIL import Image
import numpy as np

image = Image.open("../web/images/news-image.png")
print(image)
# 转换成numpy数组
image_arr = np.array(image)
print(image_arr.shape)

# 翻转图片
# image_arr[start:end:step, :, :]
# image_arr[-1::-1, :, :] 是对 image_arr（一个 3D numpy 数组，表示图片）进行切片操作
# 第一个维度 -1::-1:
# 这是图片的高度方向（即垂直轴上的行数）。
# -1 表示从最后一行开始，:: 表示按照步长翻转，-1 表示步长为 -1，即从最后一行到第一行进行翻转。因此，-1::-1 的效果是垂直方向翻转整个图像。
# 第二个维度 ::
# 这是图片的宽度方向（即水平轴上的列数）。
# 仅使用冒号 :，表示保持整个维度不变，即图片的宽度方向不会发生变化。
# 第三个维度 ::
# 这是图片的通道维度，一般对应 RGB（红、绿、蓝）三个颜色通道，或者 RGBA（多了透明度 Alpha 通道）。
# 同样使用冒号 : 表示不对通道进行任何变化，保留图片的颜色信息。
# 你可以通过切片或修改第三个维度来选择或重新排列通道。
# 获取红色通道：image_arr[:, :, 0]
# 重新排列通道：image_arr[:, :, [2, 1, 0]]
# 去掉 Alpha 通道（如果图像是 RGBA 格式）：image_arr[:, :, :3]

image_arr_up = image_arr[0:339:2,0:631:2,1]
print(image_arr_up.shape)
Image.fromarray(image_arr_up).show()
```

```python
# sigmoid 函数
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 3))
plt.rcParams["font.sans-serif"] = ["Arial"]

x = np.arange(-10, 10, 0.1)
s = 1.0 / (1 + np.exp(-x))
y = np.clip(x, a_min=0, a_max=None)
print(s)

plt.subplot(121)
plt.plot(x, s, color="r")
plt.text(-5, 0.9, r"$y=\sigma(x)$", fontsize=13)
# 设置坐标轴格式
currentAxis = plt.gca()  # 获取到当前的坐标轴
currentAxis.xaxis.set_label_text("x", fontsize=15)
currentAxis.yaxis.set_label_text("y", fontsize=15)

plt.subplot(122)
plt.plot(x, y, color="g")
plt.text(-3.0, 9, r"$y=RuLU(x)$", fontsize=13)
currentAxis = plt.gca()
currentAxis.xaxis.set_label_text("x", fontsize=15)
currentAxis.yaxis.set_label_text("y", fontsize=15)

plt.show()
```

```python
# 打车费用预测
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.title("训练轮次和损失函数分析")
plt.xlabel("训练轮次")
plt.ylabel("损失值")
plt.xticks(np.arange(0, 210, 10))
plt.yticks(np.arange(0, 2.0, 0.2))
# plt.gca().yaxis.set_major_formatter(plt.FormatStrFormatter('%.1f'))
num_epochs = 200 # 训练轮次
initial_loss = 1 # 初始损失值
decay_rate = 5 # 增加衰减速率，使初始下降更快
later_decay = 0.5 # 后期的衰减速率，使下降平滑
# 使用指数函数模拟损失下降，加入不同阶段的衰减率
t = np.linspace(0, 1, num_epochs)
# print(t)
# 设置随机种子
np.random.seed(0)
random_value = np.random.normal(0, 0.02, num_epochs)
print(random_value)
loss_values = initial_loss * np.exp(-decay_rate * t) + random_value
loss_values += np.exp(-later_decay * t) # 添加更平滑的衰减趋势

xdata = np.arange(0, 200, 1)
ydata = 1 / (1 + np.exp(-xdata))
plt.plot(loss_values, color="r")

plt.show()
```
