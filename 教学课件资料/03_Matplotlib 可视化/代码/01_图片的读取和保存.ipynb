```python
# 图片的读取和保存
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 读取图片
data = mpimg.imread('a.jpg') # data numpy的数组
print(data.shape)

# 显示图片
#  如果是黑白图片, 需要设置参数cmap='gray'
plt.imshow(data)

# 显示图片
plt.show()

# 保存图片
plt.imsave('b.jpg',data)
```

```python
# 图片处理
# 图片的读取和保存
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np 

# 读取数据
data = mpimg.imread('a.jpg')
print(data.shape)

# 1 降低图片分辨率为原来的一半
data_new = data[::2,::2]
print(data_new.shape)
# 2 把图片进行水平翻转
data_new = data_new[::1,-1::-1]


# 不显示坐标
plt.axis(False)
plt.imshow(data_new)
plt.show()
# 保存处理后的图片
plt.imsave('b.jpg',data)
```
